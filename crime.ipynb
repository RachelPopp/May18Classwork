```python
from matplotlib import pyplot as plt
from scipy.stats import linregress
import numpy as np
import pandas as pd
```

```python
crime_data = pd.read_csv("crime_data.csv", header=None)
crime_data.head()
```

```python
year = crime_data.iloc[:,0]
year.head()
violent_crime_rate = crime_data.iloc[:,3]
murder_rate = crime_data.iloc[:,5]
aggravated_assault_rate = crime_data.iloc[:,11]

violent_crime_rate.drop(violent_crime_rate.head(1).index, inplace=True)
murder_rate.drop(murder_rate.head(1).index, inplace=True)
aggravated_assault_rate.drop(aggravated_assault_rate.head(1).index, inplace=True)
year.drop(year.head(1).index, inplace=True)
```

```python
x_axis = pd.to_numeric(year)
y_axis = pd.to_numeric(violent_crime_rate)
# Set line
(slope, intercept, _, _, _) = linregress(x_axis, y_axis)
fit1 = slope * x_axis + intercept

x_axis2 = pd.to_numeric(year)
y_axis2 = pd.to_numeric(murder_rate)
# Set line
(slope, intercept, _, _, _) = linregress(x_axis2, y_axis2)
fit2 = slope * x_axis2 + intercept

x_axis3 = pd.to_numeric(year)
y_axis3 = pd.to_numeric(aggravated_assault_rate)
# Set line
(slope, intercept, _, _, _) = linregress(x_axis3, y_axis3)
fit3 = slope * x_axis3 + intercept
```

```python
# Plot data
fig, ax = plt.subplots()

fig.suptitle("Crime Rates", fontsize=16, fontweight="bold")

ax.set_xlabel("Year")
ax.set_ylabel("Rate")

ax.plot(x_axis, y_axis, linewidth=0, marker='o', label="Violent Crime Rate")
ax.plot(x_axis, fit1, 'b--')

ax.plot(x_axis2, y_axis2, linewidth=0, marker='o', label = "Murder Rate")
ax.plot(x_axis2, fit2, 'r--')

ax.plot(x_axis3, y_axis3, linewidth=0, marker='o', label="Aggravated Assault Rate")
ax.plot(x_axis3, fit3, 'g--')

ax.legend(loc='upper right', frameon=False)

plt.show()
```
